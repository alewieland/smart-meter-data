```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
from urllib.parse import urljoin
```

```python

def fetch_smartmeter_data_urls():
    base_url = "https://open.data.axpo.com/%24web/"
    url = base_url + "index.html#dataset-a"
    response = requests.get(url)
    if response.status_code != 200:
        print("Failed to fetch the webpage")
        return []

    soup = BeautifulSoup(response.content, 'html.parser')
    urls = []

    # Find all <a> tags with href containing "ckw_opendata_smartmeter_dataset_a" and ".csv.gz"
    for link in soup.find_all('a', href=re.compile(r'ckw_opendata_smartmeter_dataset_a.*\.csv\.gz')):
        urls.append(urljoin(base_url, link['href']))

    return urls

def load_dataframe_from_url(url):
    df = pd.read_csv(url)
    return df

def save_dataframe_to_parquet(df, filename):
    df.to_parquet(filename)

# Test the function

```
